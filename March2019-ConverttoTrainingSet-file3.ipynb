```python
import pandas as pd
import numpy as np
```

```python
class Game:
    
    def __init__(self,myteam,oppteam,gameid,bigdf):
        self.myteam=myteam
        self.oppteam=oppteam
        self.gameid=gameid
        self.myteam_allgamesdf=self.getmyteamAllGamesDF(bigdf)
        self.oppteam_allgamesdf=self.getoppteamAllGamesDF(bigdf)
        
    @staticmethod
    def gettime(text):
        return int(text.split(':')[0])+int(text.split(':')[1])/60
        
    def getmyteamAllGamesDF(self,allgamesdf):
        
        df=allgamesdf[(allgamesdf.hometeam==self.myteam) | (allgamesdf.awayteam==self.myteam)]
        return df.sort_values(by='gamedate',ascending=False).reset_index(drop=True)
    
    def getoppteamAllGamesDF(self,allgamesdf):
        
        df=allgamesdf[(allgamesdf.hometeam==self.oppteam) | (allgamesdf.awayteam==self.oppteam)]
        return df.sort_values(by='gamedate',ascending=False).reset_index(drop=True)
    
    def getmyteamPreviousWins(self,lookback_noofgames):

        df=self.myteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        wins=0
        for index in df.index:
            if df.loc[index,'hometeam']==self.myteam:
                if df.loc[index,'hometeam_score']>df.loc[index,'awayteam_score']:
                    wins+=1
            elif df.loc[index,'awayteam']==self.myteam:
                if df.loc[index,'awayteam_score']>df.loc[index,'hometeam_score']:
                    wins+=1
        return wins
    
    def getoppteamPreviousWins(self,lookback_noofgames):
        
        df=self.oppteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        wins=0
        for index in df.index:
            if df.loc[index,'hometeam']==self.oppteam:
                if df.loc[index,'hometeam_score']>df.loc[index,'awayteam_score']:
                    wins+=1
            elif df.loc[index,'awayteam']==self.oppteam:
                if df.loc[index,'awayteam_score']>df.loc[index,'hometeam_score']:
                    wins+=1
        return wins
    
    def getGameDate(self):
        
        return self.myteam_allgamesdf.loc[self.gameid,'gamedate']
    
    
    def getHomeOrAway(self):
        
        return self.myteam_allgamesdf.loc[self.gameid,'hometeam']==self.myteam
    
    def getGameResult(self):
        
        result=0
        df=self.myteam_allgamesdf.loc[self.gameid,:]
        if df['hometeam']==self.myteam:
            if df['hometeam_score']>df['awayteam_score']:
                result=1
            else:
                result=0
        elif df['awayteam']==self.myteam:
            if df['awayteam_score']>df['hometeam_score']:
                result=1
            else:
                result=0
        return result
    
    def getmyteamScores(self,lookback_noofgames):
        
        df=self.myteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        scores=0
        for index in df.index:
            if df.loc[index,'hometeam']==self.myteam:
                scores+=df.loc[index,'hometeam_score']
            elif df.loc[index,'awayteam']==self.myteam:
                scores+=df.loc[index,'awayteam_score']
        return scores
    
    
    def getoppteamScores(self,lookback_noofgames):
        
        df=self.oppteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        scores=0
        for index in df.index:
            if df.loc[index,'hometeam']==self.oppteam:
                scores+=df.loc[index,'hometeam_score']
            elif df.loc[index,'awayteam']==self.oppteam:
                scores+=df.loc[index,'awayteam_score']
        return scores
    
    def getmyteamTrnOvr(self,lookback_noofgames):
        
        df=self.myteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        trnovrs=0
        for index in df.index:
            if df.loc[index,'hometeam']==self.myteam:
                trnovrs+=df.loc[index,'homestats_trnovr']
            elif df.loc[index,'awayteam']==self.myteam:
                trnovrs+=df.loc[index,'awaystats_trnovr']
        return trnovrs
    
    def getoppteamTrnOvr(self,lookback_noofgames):
        
        df=self.oppteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        trnovrs=0
        for index in df.index:
            if df.loc[index,'hometeam']==self.oppteam:
                trnovrs+=df.loc[index,'homestats_trnovr']
            elif df.loc[index,'awayteam']==self.oppteam:
                trnovrs+=df.loc[index,'awaystats_trnovr']
        return trnovrs
    
    def getmyteamTop(self,lookback_noofgames):
    
        df=self.myteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        timeofpos=0
        for index in df.index:
            if df.loc[index,'hometeam']==self.myteam:
                timeofpos+=Game.gettime(df.loc[index,'homestats_top'])
            elif df.loc[index,'awayteam']==self.myteam:
                timeofpos+=Game.gettime(df.loc[index,'awaystats_top'])
        return timeofpos
    
    def getoppteamTop(self,lookback_noofgames):
        
        df=self.oppteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        timeofpos=0
        for index in df.index:
            if df.loc[index,'hometeam']==self.oppteam:
                timeofpos+=Game.gettime(df.loc[index,'homestats_top'])
            elif df.loc[index,'awayteam']==self.oppteam:
                timeofpos+=Game.gettime(df.loc[index,'awaystats_top'])
        return timeofpos
    
    def getmyteamYds(self,ydstype,lookback_noofgames):
    
        df=self.myteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        yds=0
        for index in df.index:
            if ydstype=='rush':
                if df.loc[index,'hometeam']==self.myteam:
                    yds+=df.loc[index,'homestats_ryds']
                elif df.loc[index,'awayteam']==self.myteam:
                    yds+=df.loc[index,'awaystats_ryds']
            
            elif ydstype=='pass':
                if df.loc[index,'hometeam']==self.myteam:
                    yds+=df.loc[index,'homestats_pyds']
                elif df.loc[index,'awayteam']==self.myteam:
                    yds+=df.loc[index,'awaystats_pyds']
            
            elif ydstype=='punt':
                if df.loc[index,'hometeam']==self.myteam:
                    yds+=df.loc[index,'homestats_ptyds']
                elif df.loc[index,'awayteam']==self.myteam:
                    yds+=df.loc[index,'awaystats_ptyds']
                    
            elif ydstype=='pen':
                if df.loc[index,'hometeam']==self.myteam:
                    yds+=df.loc[index,'homestats_penyds']
                elif df.loc[index,'awayteam']==self.myteam:
                    yds+=df.loc[index,'awaystats_penyds']
        
        return yds
    
    
    def getoppteamYds(self,ydstype,lookback_noofgames):
        
        df=self.oppteam_allgamesdf
        df=df.loc[self.gameid+1:self.gameid+lookback_noofgames,:]
        yds=0
        for index in df.index:
            if ydstype=='rush':
                if df.loc[index,'hometeam']==self.oppteam:
                    yds+=df.loc[index,'homestats_ryds']
                elif df.loc[index,'awayteam']==self.oppteam:
                    yds+=df.loc[index,'awaystats_ryds']
            
            elif ydstype=='pass':
                if df.loc[index,'hometeam']==self.oppteam:
                    yds+=df.loc[index,'homestats_pyds']
                elif df.loc[index,'awayteam']==self.oppteam:
                    yds+=df.loc[index,'awaystats_pyds']
            
            elif ydstype=='punt':
                if df.loc[index,'hometeam']==self.oppteam:
                    yds+=df.loc[index,'homestats_ptyds']
                elif df.loc[index,'awayteam']==self.oppteam:
                    yds+=df.loc[index,'awaystats_ptyds']
                    
            elif ydstype=='pen':
                if df.loc[index,'hometeam']==self.oppteam:
                    yds+=df.loc[index,'homestats_penyds']
                elif df.loc[index,'awayteam']==self.oppteam:
                    yds+=df.loc[index,'awaystats_penyds']
        
        return yds
    
    
    
# game=Game('PHI','DAL',3)
# game.getmyteamYds('rush',3),game.getoppteamYds('pass',3)
```

```python
def GetTrainingDF(lookbacknoofgames,bigdf):
    
    teamlist=list(set(bigdf.hometeam.values))
    
    mytrainingdf={}
    mytrainingdf['myteam']=[]
    mytrainingdf['oppteam']=[]
    mytrainingdf['myteam_prvwins']=[]
    mytrainingdf['oppteam_prvwins']=[]
    mytrainingdf['homegame']=[]
    mytrainingdf['result']=[]
    mytrainingdf['myteam_scores']=[]
    mytrainingdf['oppteam_scores']=[]
    mytrainingdf['myteam_trnovr']=[]
    mytrainingdf['oppteam_trnovr']=[]
    mytrainingdf['myteam_top']=[]
    mytrainingdf['oppteam_top']=[]
    mytrainingdf['myteam_pyds']=[]
    mytrainingdf['oppteam_pyds']=[]
    mytrainingdf['myteam_ryds']=[]
    mytrainingdf['oppteam_ryds']=[]
    mytrainingdf['myteam_penyds']=[]
    mytrainingdf['oppteam_penyds']=[]
    mytrainingdf['myteam_ptyds']=[]
    mytrainingdf['oppteam_ptyds']=[]
    mytrainingdf['gamedate']=[]


    for team in teamlist:
        df=bigdf[(bigdf.hometeam==team) | (bigdf.awayteam==team)].sort_values(by='gamedate',ascending=False).reset_index(drop=True)
        oppteams=[df.loc[index,'hometeam'] if df.loc[index,'hometeam']!=team else df.loc[index,'awayteam'] for index in df.index ]
        gameids=df.index

        for gameid,myteam,oppteam in list(zip(gameids,[team]*len(oppteams),oppteams))[:-lookbacknoofgames]:
            game=Game(myteam,oppteam,gameid,bigdf)
            mytrainingdf['myteam'].append(game.myteam)
            mytrainingdf['oppteam'].append(game.oppteam)
            mytrainingdf['myteam_prvwins'].append(game.getmyteamPreviousWins(lookbacknoofgames))
            mytrainingdf['oppteam_prvwins'].append(game.getoppteamPreviousWins(lookbacknoofgames))
            mytrainingdf['gamedate'].append(game.getGameDate())
            mytrainingdf['homegame'].append(game.getHomeOrAway())
            mytrainingdf['result'].append(game.getGameResult())
            mytrainingdf['myteam_scores'].append(game.getmyteamScores(lookbacknoofgames))
            mytrainingdf['oppteam_scores'].append(game.getoppteamScores(lookbacknoofgames))
            mytrainingdf['myteam_trnovr'].append(game.getmyteamTrnOvr(lookbacknoofgames))
            mytrainingdf['oppteam_trnovr'].append(game.getoppteamTrnOvr(lookbacknoofgames))
            mytrainingdf['myteam_top'].append(game.getmyteamTop(lookbacknoofgames))
            mytrainingdf['oppteam_top'].append(game.getoppteamTop(lookbacknoofgames))
            mytrainingdf['myteam_pyds'].append(game.getmyteamYds('pass',lookbacknoofgames))
            mytrainingdf['oppteam_pyds'].append(game.getoppteamYds('pass',lookbacknoofgames))
            mytrainingdf['myteam_ryds'].append(game.getmyteamYds('rush',lookbacknoofgames))
            mytrainingdf['oppteam_ryds'].append(game.getoppteamYds('rush',lookbacknoofgames))
            mytrainingdf['myteam_penyds'].append(game.getmyteamYds('pen',lookbacknoofgames))
            mytrainingdf['oppteam_penyds'].append(game.getoppteamYds('pen',lookbacknoofgames))
            mytrainingdf['myteam_ptyds'].append(game.getmyteamYds('punt',lookbacknoofgames))
            mytrainingdf['oppteam_ptyds'].append(game.getoppteamYds('punt',lookbacknoofgames))

    return pd.DataFrame(mytrainingdf)
```

```python
lookbacknoofgames=4
df2018=pd.read_pickle('games2018seasondf.pickle')
df2017=pd.read_pickle('games2017seasondf.pickle')

trainingset2018=GetTrainingDF(lookbacknoofgames,df2018)
trainingset2017=GetTrainingDF(lookbacknoofgames,df2017)

bigtrainingdataset=trainingset2018.append(trainingset2017)
```

```python
def buildkey(df):
    key=sorted((df['myteam'],df['oppteam']))
    key.append(df['gamedate'].strftime('%Y/%m/%d'))
    return '-'.join(key)

bigtrainingdataset['key']=bigtrainingdataset.apply(buildkey,axis=1)
bigtrainingdataset=bigtrainingdataset.sort_values(by='key').drop_duplicates(subset=['key'], keep='first')
```
